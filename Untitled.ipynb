```python
import os
import json
import pandas as pd

current_dir = os.path.dirname(os.getcwd())
data_dir=os.path.join(current_dir,'metis_project1','data','metacritic')

metacritic_data_raw=[]

for file in os.listdir(data_dir):
        target_file_path = os.path.join(data_dir, file)
        with open(target_file_path, 'r') as target_file:
            json_file=json.load(target_file)
            if isinstance(json_file,dict):
                metacritic_data_raw.append(json_file)
            
mdata = pd.DataFrame(metacritic_data_raw)
```
